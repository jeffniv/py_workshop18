## Pandas and Seaborn

This is workshop will focus on learning `pandas` and `seaborn` packages. These two packages are the most widely used Python packages for Data Analytics.

```python
import sys
sys.version
```

```python
%matplotlib inline
import pandas as pd
import numpy as np
import seaborn as sns
```

Pandas is the most widely used package for Data Analytics. It puts the data into a "tidy" format, similar to R, and incorporates basic functions for statistical analysis and visualization. There are two main types of data structures in pandas: Series and DataFrames. We will focus mainly on DataFrames, though we can view the columns of the DataFrames as individual Series, which incorporate many of the DataFrame methods.

```python
iris = sns.load_dataset('iris')
iris
```

The columns of a `DataFrame` are `Series` objects.

```python
iris['sepal_length']
```

There are many different methods for slicing a DataFrame.

```python
iris.iloc[:,1:3]
```

```python
iris[iris.species == 'setosa']
```

```python
We can use the `query` method to slice a DataFrame according to multiple conditions
```

```python
iris.query('species == "setosa" & sepal_length < 5.0 & sepal_width < 3')
```

We can get basic descriptive statistics using `DataFrame.describe()`.

```python
iris.describe()
```

There are other useful functions that can be called on the pandas object.

```python
iris.corr()
```

Pandas comes with some nifty visualization methods. The `DataFrame.plot()` method allows the user to choose what `kind` of plot to be rendered.

```python
iris.plot(kind='kde')
```

There are also some more advanced visualization methods under the `pandas.plotting` module.

```python
pd.plotting.parallel_coordinates(iris, 'species')
```

```python
iris.boxplot('sepal_length', by='species')
```

Pandas incorporates easily with matplotlib for options such as layout, style, etc.

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
fig, [ax1,ax2] = plt.subplots(1,2, figsize=(10,5))
iris.plot(kind='kde', ax=ax1)
pd.plotting.parallel_coordinates(iris, 'species', ax=ax2)
plt.show()
```

`seaborn` is a visualization package that is similar to the popular `ggplot2` used in `R`. There are a couple plug-n-play functions in the base package to visualize relationships within data.

```python
sns.pairplot(data=iris, hue='species')
```

```python
sns.jointplot(x='sepal_length', y='petal_length', data=iris, kind='hex')
```

We can visualize a linear regression on the data using the function `lmplot`. Seaborn mainly utilizes the `statsmodels` package for its statistical analysis.

```python
reg = sns.lmplot(x='sepal_length', y='petal_length', data=iris, hue='species' )
```
